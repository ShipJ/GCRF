```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import sklearn.metrics
```

Retrieve master data set

```python
data = pd.DataFrame(pd.read_csv('data/processed/civ/master_pop.csv'))
data['Vol_pp'] = data['Vol']/data['Pop_2010']
data['Pop_density'] = data['Area(km^2)'] / data['Pop_2010']
data = data[data['Adm_1'] != 1]
data = data[data['Adm_2'] != 2]
data = data[data['HIVPosRate'] != 0]
```

```python
X = data[['Pop_density', 'DeathRate']]

def normalise(df):
    return (df - df.mean()) / df.std()

X = normalise(X).as_matrix()
mae = []
for i in range(5):
    mae_1000 = []
    train = 40+(i*9)
    for j in range(2000):
        np.random.shuffle(X)
        X_train = X[0:train, [0]]
        y_train = X[0:train, 1]
        X_test = X[train:, [0]]
        y_test = X[train:, 1]

        lm = LinearRegression()
        lm.fit(X_train, y_train)

        mae_1000.append(np.mean(np.abs(lm.predict(X_test) - y_test)))
    mae.append(np.median(mae_1000))
    
plt.plot(range(5), mae, '-x', c='b', label='Pop_1km')
```

```python
X = data[['Pop_density', 'Vol_pp', 'Introversion', 'EigenvectorCentrality',
          'Degree_centrality', 'Med_degree', 'DeathRate']]

def normalise(df):
    return (df - df.mean()) / df.std()

X = normalise(X).as_matrix()
mae = []
for i in range(5):
    mae_1000 = []
    train = 40+(i*9)
    for j in range(2000):
        np.random.shuffle(X)
        X_train = X[0:train, [0,1,2,3,4,5]]
        y_train = X[0:train, 6]
        X_test = X[train:, [0,1,2,3,4,5]]
        y_test = X[train:, 6]

        lm = LinearRegression()
        lm.fit(X_train, y_train)
    
    mae.append(np.median(mae_1000))

plt.plot(range(5), mae, '-x', c='r', label='Pop_1km + vol')
plt.show()
```

```python
X = data[['Pop_1km', 'Vol_pp', 'Introversion', 'DeathRate']]

def normalise(df):
    return (df - df.mean()) / df.std()

X = normalise(X).as_matrix()


mae = []
for i in range(5):
    mae_1000 = []
    train = 70+(i*13)
    for j in range(5000):
        np.random.shuffle(X)
        X_train = X[0:train, [0,1,2]]
        y_train = X[0:train, 3]
        X_test = X[train:, [0,1,2]]
        y_test = X[train:, 3]

        lm = LinearRegression()
        lm.fit(X_train, y_train)
        mae_1000.append(np.mean(np.abs(lm.predict(X_test) - y_test)))
    mae.append(np.mean(mae_1000))
plt.plot(range(5), mae, '-x', c='g', label='Pop_1km + vol + ent')
plt.legend()
plt.show()
```
